## FIR filter implementation

```python
import numpy as np 
#FIR filter formular: y(n)=âˆ‘h(m)x(n-m)

# h is the coefficients generated by the filterdesign of MATLAB
h = [1,2,3]
# x is the signal we want to filte 
x = [0,1,0.5]
# linear convolution implemented by convolve method
y = np.convolve(h,x)
print(y)
```

## IIR filter implementation

```python
import copy
#IIR filter formular: y(n) = b(1)*x(n) + b(2)*x(n-1) + ... + b(Nb+1)*x(n-nb) - a(2)*y(n-1) - ... - a(Na+1)*y(n-na)
a_EEG = [1, -1.63673461004437, 0.639034958513405]
b_EEG = [0.180482520743298, 0, -0.180482520743298]

in1 = []
out1 = []
for i in range(len(b_EEG)):
    in1.append(0)
for i in range(len(a_EEG)-1):
    out1.append(0)
    
def IIRFilter(signal,a,b):
    outData = []
    for i in range(len(signal)):
        outData.append(0)
    
    for i in range(len(signal)):
        temp = copy.deepcopy(in1) 
        for j in range(len(in1)-1):
            in1[j+1] = temp[j]
        in1[0] = signal[i]
        
        y = 0
        for j in range(len(b)):
            y += b[j]*in1[j]
        for j in range(len(a)-1):
            y -= a[j+1]*out1[j]
            
        temp_out1 = copy.deepcopy(out1) 
        for j in range(len(out1)-1):
            out1[j+1] = temp_out1[j]
        out1[0] = y
        outData[i] = y
    
    return outData


y2 = IIRFilter(x,a_EEG,b_EEG)
print(y2)
```
